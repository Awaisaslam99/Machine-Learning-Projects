```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with Friends forever 👑👑👑.txt', encoding= 'utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
# Regular expression pattern for dates
date_pattern = re.compile(r'(\d{1,2}/\d{1,2}/\d{4},\s\d{1,2}:\d{2}\s[APMapm]{2})')

# Extract dates from each message
dates = [date.replace('\u202f', '') for date in date_pattern.findall(data)]

# Split the data by newline characters to create the 'messages' list
details_pattern = re.compile(r'\d{1,2}/\d{1,2}/\d{4},\s\d{1,2}:\d{2}\s[APMapm]{2} - (\w+\s\w+): (.*)')
# Extract details (name and message) from each message
messages = []
matches = details_pattern.findall(data)
for match in matches:
    messages.append(match)
# Match the length of 'dates' and 'messages'
min_length = min(len(dates), len(messages))
dates = dates[:min_length]
messages = messages[:min_length]
```

```python
# Create DataFrame
df = pd.DataFrame({'user_message': messages, 'message_date': dates})


df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%Y, %I:%M%p', errors='coerce')

df.rename(columns= {'message_date':'date'}, inplace=True)
```

```python
names = []
messages = []

for message in df['user_message']:
    entry = re.match(r"\('([^']+)', '(.*)'\)", str(message))
    if entry:
        names.append(entry.group(1).strip())
        messages.append(entry.group(2).strip())
    else:
        names.append('group_notification')
        messages.append(str(message).strip())

df['user'] = names
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)

# Display the resulting DataFrame
print(df)
```
